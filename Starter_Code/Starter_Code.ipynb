## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.

app_df = application_df.drop(['EIN', 'NAME'], axis=1)
```

```python
# Determine the number of unique values in each column.
counts = app_df.nunique()
print(counts)
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_type_val = app_df['APPLICATION_TYPE'].value_counts()
app_type_val
```

```python
# separate values to be replaced
application_types_to_replace = ["T9", "T13", "T12", "T2", "T25", "T14", "T29", "T15", "T17"]
```

```python
# Replace in dataframe
for app in application_types_to_replace:
    app_df['APPLICATION_TYPE'] = app_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
app_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
class_val_counts = app_df['CLASSIFICATION'].value_counts()
class_val_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
filtered_counts = class_val_counts[class_val_counts > 1]
filtered_counts
```

```python
classifications_to_replace = class_val_counts[class_val_counts < 1883]

# Replace in dataframe
for cls in classifications_to_replace.index:  # Iterate over index labels
    app_df['CLASSIFICATION'] = app_df['CLASSIFICATION'].replace(cls, "Other")

# Check to make sure binning was successful
print(app_df['CLASSIFICATION'].value_counts())
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
app_df = pd.get_dummies(app_df)
app_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = app_df['IS_SUCCESSFUL']
X = app_df.drop(columns='IS_SUCCESSFUL')

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE

nn = tf.keras.models.Sequential()
nn.add(tf.keras.layers.Dense(units=80, activation="relu", input_dim=43))


# First hidden layer
nn.add(tf.keras.layers.Dense(units=30, activation="relu"))

# Second hidden layer


# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# call back to save models weights every 5 epochs
class SaveWeightsCallback(tf.keras.callbacks.Callback):
    def __init__(self, save_path):
        super(SaveWeightsCallback, self).__init__()
        self.save_path = save_path
        self.epoch_count = 0

    def on_epoch_end(self, epoch, logs=None):
        if (epoch + 1) % 5 == 0:  # Save weights every 5 epochs
            self.model.save_weights(f"{self.save_path}_epoch_{epoch+1}.h5")
            print(f"Saved weights at epoch {epoch+1}")


save_callback = SaveWeightsCallback(save_path="model_weights")
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100, callbacks=[save_callback])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
